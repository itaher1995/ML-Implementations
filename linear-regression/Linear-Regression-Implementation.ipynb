## Closed Form Linear Regression

```python
def analyticalLinearRegressionFit(X,Y,mapping):
    """
    An implementation of the closed form solution of linear regression. 
    Allows for nonlinear mapping in order to better fit data.
    """
    if len(X)>len(X):
        return "Can't compute closed form solution for Linear Regression"
    if mapping<1:
        return "Mapping is sublinear"
    else:
        X_copy = np.copy(X) #create a copy of the data
        X_copy = np.array([np.array([e**d for d in range(mapping+1) for e in x]) for x in X_copy]) #basis function expansion
        theta = np.dot(np.linalg.inv(np.dot(X_copy.T,X_copy)),np.dot(X_copy.T,Y)) #computes closed form solution (X^TX)^-1(X^TY)
    return theta

def analyticalLinearRegressionTest(X,theta,mapping):
    """
    An implementation of a testing function to develop predictions using trained theta based 
    on the closed form solution.
    """
    if len(X)>len(X):
        return "Can't compute closed form solution for Linear Regression"
    if mapping<1:
        return "Mapping is sublinear"
    else:
        X_copy = np.copy(X) #create a copy of the data
        X_copy = np.array([np.array([e**d for d in range(mapping+1) for e in x]) for x in X_copy])
        y_pred = np.dot(X_copy,theta)
    return y_pred
```

## Stochastic Gradient Descent

```python
def linearRegressionStochasticGradientDescentFit(X,Y,m,alpha,mapping,num_iters):
    """
    Implementation of linear regression using the stochastic gradient descent heuristic.
    X and Y correspond to the training data, m is the batch size, alpha is the learning rate, 
    mapping is the the value used for the basis function expansion to allow for nonlinear mapping.
    Finally num_iters is how many iterations are done for convergence.
    """
    if mapping<1:
        return "Mapping is sublinear"
    
    X_copy = np.copy(X) #create a copy of the data
    X_copy = np.array([np.array([e**d for d in range(mapping+1) for e in x]) for x in X_copy]) #basis function expansion
    
    for i in range(len(X_copy[0])):
        #for all col i in X get the mean and standard deviation
        col=np.array([[X_copy[j][i]] for j in range(len(X_copy))])
        col_mean=np.mean(col) 
        col_std = np.std(col)
        for j in range(len(X_copy)):
            if col_std!=0:
                X_copy[j][i]=(X_copy[j][i]-col_mean)/col_std #scale data to prevent theta overflow

    
    theta = np.array([np.array([0])]*len(X_copy[0])) #initialize theta to be 0 vector for hard start.
    gradient = np.array([np.array([1000])]*len(X_copy[0]))
    
    epsilon=10**-4
    k=0
    while np.dot(gradient.T,gradient)>=epsilon: #repeat until convergence (gradient < epsilon)
        m_indices = random.sample(range(len(X_copy)),m) #grab m values at m random indices for stochastic gradient descent
        gradient = alpha * np.dot(X_copy[m_indices].T,np.dot(X_copy[m_indices],theta)-Y[m_indices]) #compute new theta using subset (good for space complexity AND time complexity)
        theta = theta - gradient
        k+=1
    print(k)
    return theta

def linearRegressionStochasticGradientDescentTest(X,theta,mapping):
    if mapping<1:
        return "Mapping is sublinear"
    else:
        X_copy = np.copy(X) #create a copy of the data
        X_copy = np.array([np.array([e**d for d in range(mapping+1) for e in x]) for x in X_copy])
        for i in range(len(X_copy[0])):
            #for all col i in X get the mean and standard deviation
            col=np.array([[X_copy[j][i]] for j in range(len(X_copy))])
            col_mean=np.mean(col) 
            col_std = np.std(col)
            for j in range(len(X_copy)):
                if col_std!=0:
                    X_copy[j][i]=(X_copy[j][i]-col_mean)/col_std #scale data to prevent theta overflow
                else:
                    X_copy[j][i]-(X_copy[j][i]-col_mean)
        y_pred = np.dot(X_copy,theta)
    return y_pred

        
```

## Closed Form w/ Regularization

```python
def analyticalLinearRegressionRegularizedFit(X,Y,mapping,lambda_):
    if len(X)>len(X):
        return "Can't compute closed form solution for Linear Regression"
    if mapping<1:
        return "Mapping is sublinear"
    else:
        X_copy = np.copy(X) #create a copy of the data
        X_copy = np.array([np.array([e**d for d in range(mapping+1) for e in x]) for x in X_copy]) #basis function expansion
        theta = np.dot(np.linalg.inv(np.dot(X_copy.T,X_copy)-lambda_*np.identity(n=len(X_copy[0]))),np.dot(X_copy.T,Y))
    return theta
```

## Stochastic Gradient Descent w/ Regularization

```python
def linearRegressionStochasticGradientDescentRegularizedFit(X,Y,m,alpha,mapping,lambda_):
    if mapping<1:
        return "Mapping is sublinear"
    
    X_copy = np.copy(X) #create a copy of the data
    X_copy = np.array([np.array([e**d for d in range(mapping+1) for e in x]) for x in X_copy]) #basis function expansion
    
    for i in range(len(X_copy[0])):
        #for all col i in X get the mean and standard deviation
        col=np.array([[X_copy[j][i]] for j in range(len(X_copy))])
        col_mean=np.mean(col) 
        col_std = np.std(col)
        for j in range(len(X_copy)):
            if col_std!=0:
                X_copy[j][i]=(X_copy[j][i]-col_mean)/col_std #scale data to prevent theta overflow

    
    theta = np.array([np.array([0])]*len(X_copy[0])) #initialize theta to be 0 vector for hard start.
    gradient = np.array([np.array([1000])]*len(X_copy[0]))
    
    epsilon=10**-4
    
    while np.dot(gradient.T,gradient)>=epsilon: #repeat until convergence (gradient < epsilon)
        m_indices = random.sample(range(len(X_copy)),m) #grab m values at m random indices for stochastic gradient descent
        gradient = alpha * 2*np.dot(X_copy[m_indices].T,np.dot(X_copy[m_indices],theta)-Y[m_indices]) + 2*lambda_*theta
        theta = theta - gradient
    return theta
        
```

## Error Function 

$\Sigma ||predictions - actual||_{2}^2$

```python
def SSE(predictions,actual):
    return sum((predictions-actual)**2)
```

## Question 6b.

### Closed-form

```python
minibatch_speed_2_CF=[]
minibatch_error_2_CF=[]
for n in [2,3,5]:
    for m in [10,20,30]:
        start = time.time()
        theta=analyticalLinearRegressionFit(x_train1, y_train1,n)
        y_pred_train = analyticalLinearRegressionTest(x_train1, theta, n)
        y_pred_test = analyticalLinearRegressionTest(x_test1, theta, n)
        end = time.time()
        print("Theta: {0}".format(theta))
        print("Training error using Sum of Square Errors: {0}".format(SSE(y_pred_train,y_train1)))
        print("Testing error using Sum of Square Errors: {0}".format(SSE(y_pred_test,y_test1)))
        minibatch_speed_2_CF.append(end-start)
        minibatch_error_2_CF.append(SSE(y_pred_test,y_test1))
```

```python
#plot for speed
fig = plt.figure()

ax = fig.add_subplot(111)
plt.plot(minibatch_speed_2_CF)
ax.set_xticklabels([])
plt.title("Change in Speed for Closed Form as m increases")
```

```python
#plot for error
fig = plt.figure()

ax = fig.add_subplot(111)
plt.plot(minibatch_error_2_CF)
ax.set_xticklabels([])
plt.title("Change in Error for Closed Form as m increases")
```

## Question 6c.

```python
def k_foldCV(X,Y,folds):
    lenFolds = len(X)//folds
    xFolds = [X[i*lenFolds:i*lenFolds+lenFolds] for i in range(folds)]
    yFolds = [Y[i*lenFolds:i*lenFolds+lenFolds] for i in range(folds)]
    return (xFolds,yFolds)

```

### Closed Form Hyperparameter Tuning

```python
lambdas = [0.001, 0.005, 0.01]
n = [2,3,5]
K = [2,10,len(x_train2)]


for k in K:
    print(k)
    xfold, yfold=k_foldCV(x_train2,y_train2,k)
    for i in range(len(xfold)):
        xfold_copy = xfold[:]
        yfold_copy = yfold[:]
        hold_out_X=xfold_copy[i]
        hold_out_Y=yfold_copy[i]
        xfold_copy.pop(i)
        yfold_copy.pop(i)
        train_X=np.array(list(itertools.chain.from_iterable(xfold_copy)))
        train_Y=np.array(list(itertools.chain.from_iterable(yfold_copy)))
        for j in n:
            print(j)
            for lambda_ in lambdas:
                print(lambda_)
                theta = analyticalLinearRegressionRegularizedFit(train_X, train_Y, j, lambda_)
                y_pred_train = analyticalLinearRegressionTest(train_X,theta,j)
                y_pred_test = analyticalLinearRegressionTest(x_test2,theta,j)
                print("Theta: {0}".format(theta))
                print("Training error using Sum of Square Errors: {0}".format(SSE(y_pred_train,train_Y)))
                print("Testing error using Sum of Square Errors: {0}".format(SSE(y_pred_test,y_test2)))
```
